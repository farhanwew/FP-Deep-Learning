```python
%pip install ultralytics -q

# Impor library yang akan digunakan
import os
from ultralytics import YOLO
import torch

# Verifikasi aktivasi GPU
print("\nSetup Selesai.")
if torch.cuda.is_available():
    print(f"GPU terdeteksi: {torch.cuda.get_device_name(0)}")
else:
    print("GPU tidak terdeteksi, training akan berjalan di CPU (lebih lambat).")
```

```python
# Baris ini diasumsikan sudah Anda jalankan dan berhasil:
# project = rf.workspace("ctufinalthesis").project("drowsinessdetectionyolov8-test2")
# version = project.version(9)
# dataset = version.download("yolov8")

print("Persiapan training...")

# Dapatkan path ke file data.yaml yang sekarang sudah ada
data_yaml_path = os.path.join(dataset.location, 'data.yaml')
print(f"File konfigurasi ditemukan di: {data_yaml_path}")

# Muat model pre-trained YOLOv8s
print("Memuat model YOLOv8s...")
model = YOLO('yolov8s.pt')

# Mulai proses training!
print("\nMemulai training model. Ini akan berjalan tanpa error 'file not found'.")
results = model.train(
    data=data_yaml_path,
    epochs=20,
    imgsz=480,
    project="Drowsiness_Detection_Results",
    name="successful_run_1",
    exist_ok=True
)

print("\nSelamat! Proses training telah selesai.")
```

```python
import cv2
import matplotlib.pyplot as plt
print("\nMenampilkan hasil visualisasi training...")

# Direktori tempat hasil disimpan oleh `train()`
# Objek `results` memiliki atribut `save_dir` yang menunjuk ke lokasi ini
# Ini adalah cara yang lebih andal daripada membuat path secara manual
run_directory = results.save_dir

# Daftar file gambar hasil yang umum dihasilkan
result_images = {
    "Kurva Hasil (Metrik & Loss)": os.path.join(run_directory, "results.png"),
    "Confusion Matrix": os.path.join(run_directory, "confusion_matrix.png"),
    "Prediksi Batch Validasi 0 (Hasil Prediksi)": os.path.join(run_directory, "val_batch0_pred.jpg"),
    "Prediksi Batch Validasi 0 (Label Asli)": os.path.join(run_directory, "val_batch0_labels.jpg")
}

# Fungsi untuk menampilkan gambar dengan judul
def display_image(title, path):
    if os.path.exists(path):
        # Baca gambar menggunakan OpenCV dan konversi dari BGR ke RGB untuk Matplotlib
        img = cv2.imread(path)
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        plt.figure(figsize=(10, 8))
        plt.imshow(img_rgb)
        plt.title(title, fontsize=16)
        plt.axis('off') # Sembunyikan sumbu
        plt.show()
    else:
        print(f"File tidak ditemukan: {path}")

# Tampilkan setiap gambar hasil
for title, path in result_images.items():
    display_image(title, path)

print("\nVisualisasi selesai.")
```
